```python
"Итерируемые объекты и генераторы"

def geneRange(fr = 0, to = 0, step = 1):
    "The analogy of range() function"
    
    if fr and not to:
        to, fr = fr, 0
    
    while fr <= to:
#         return fr
        "and_1"
        yield fr
        fr += step
        
# print(geneRange(1, 11, 2))
"and_1"
# for i in geneRange(1, 11, 2):
#     print(i)
"or"
rang = [i for i in geneRange(1, 100) if i%10 == 7]
print(rang)

print('\n')

p = geneRange(1, 11)
print(next(p))
print(next(p))
print(next(p))
print(next(p))

##############################################
print('\n')
import random

l = [random.randint(10, 100) for _ in range(10)]
for val in enumerate(l):
    print(val)

print('\n')

for idx, val in enumerate(l):    
    print(idx,'-', val)
```

```python
"Случайности"

import random as rd
#'seed, random, randint, randrange, shuffle, choice, sample'

l = [i for i in range(10)]
rd_l = [rd.randint(10, 100) for _ in range(10)]

#'The point of start to calculate random data, in default = to current data'
# random.seed(10)

#'in interval betweene 0 & 1'
print('random - ', rd.random())

#'random int number in interval betweene 10 & 100'
print('randint - ', rd.randint(10, 100))

#''
print('randrange - ', rd.randrange(10, 100))

#''
print('until shuffle - ', l)
rd.shuffle(l)
print('after shuffle - ', l)

#'random choice one element from list'
print('choice - ', rd.choice(l))

#'random count sample(bit) of list WITHOUT repeat elements'
print('sample(WITHOUT repeat elements) - ', rd.sample(l, 3))

#'random count sample(bit) of list WITH repeat elements'
print('sample(WITH repeat elements) - ', [rd.choice(l) for _ in range(3)])
```

```python
"Регулярные выражения"

import re
#'match, search, split, sub'

some_text = """Don’t worry too much see about understanding the underlying math in this code. 
            Just note that the function takes an input number, reverses it, 
            and checks to see if the reversed number is the same as the original."""

#cheking to match the first element 
print('match_D - ', re.match('D', some_text))
print('match_i - ', re.match('i', some_text))
print('\n')

#search first matches in text
print('search_a - ', (re.search(r'a', some_text)))
print('search_x - ', re.search(r'x', some_text))
print('\n')

print('split - ', re.split('\n', some_text))

print('\n')
print('sub - ', re.sub('\t', '', some_text))
print('\n')
print('the original text - ', some_text)

```
