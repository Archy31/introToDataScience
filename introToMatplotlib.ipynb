```python
variance = [2**i for i in range(10)]
bias_squared = list(reversed(variance))

total_errors = [i+j for i, j in zip(variance, bias_squared)]

xs = [i for i in range(len(variance))]


plt.plot(xs, variance, 'g-', label='дисперсия')
plt.plot(xs, bias_squared, 'r-.', label='смещение^2')
plt.plot(xs, total_errors, 'b:', label='суммарная ошибка')

plt.legend(loc=9)

plt.xlabel('сложность модели')
plt.title('Компромис между смещением и дисперсией')

plt.show()
```

```python
friends = [random.randint(50, 97) for i in range(10)]
minutes = [i*2-23 for i in friends]
random.seed(10)

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'j', 'h', 'i', 'g']

plt.scatter(friends, minutes)
plt.axis('equal')
for label, friend, minute in zip(labels, friends, minutes):
    plt.annotate(
        label,
        xy=(friend, minute),
        xytext = (5, -5),
        textcoords = 'offset points'
    )
    
plt.show()
```
